```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
dt = pd.read_csv('mushrooms.csv')
print(dt.shape)
dt.head(10)
```

## Encoding into numerical data



```python
le = LabelEncoder()
# le.fit_transform?

ds = dt.apply(le.fit_transform)
ds.head()
```

```python
data  = ds.values

print(data[:5,:])

Y = data[:,0]
X = data[:,1:]
print(X.shape)
print(Y.shape)
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.2)
```

```python
print(x_train.shape)
print(y_train.shape)
```

## Classifier

```python
def prior_prob(y_train,label):
    
    tot = y_train.shape[0]
    num = np.sum(y_train==label)
    
    return num/float(tot)


```

```python
def conditional_prob(x_train,y_train,feat_col,feat_val,label):
    
    
    x_filterd = x_train[y_train==label]
    
    numerator = np.sum(x_filterd[:,feat_col]==feat_val)
    
    denom = np.sum(y_train==label)
    
    return numerator/float(denom)
```

## Posterior Probability

```python
def predict(x_train,y_train,xtest):
    
    posts=[]
    n_features = x_train.shape[1]
    
    for ix in np.unique(y_train):
        likelihood=1.0
        for iy in range(n_features): 
            likelihood *= conditional_prob(x_train,y_train,iy,xtest[iy],ix) 
       
        prior =  prior_prob(y_train,ix)
        post = likelihood*prior
        posts.append(post)
    
    return np.argmax(posts)
                
```

## Testing

```python
def score(x_train,y_train,x_test,y_test):
    
    
    predictions = []
    for ix in range(x_test.shape[0]):
        
        pred = predict(x_train,y_train,x_test[ix])
        predictions.append(pred)
    
    
    predictions=np.array(predictions)
   
    ans = (np.sum(predictions==y_test))/float(x_test.shape[0])
    
    return ans
```

```python
# x_train.shape
print(score(x_train,y_train,x_test,y_test))
# print(y_test.shape)
```
